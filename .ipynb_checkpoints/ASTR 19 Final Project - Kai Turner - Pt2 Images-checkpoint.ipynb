# ASTR 19 Final Pt. 2 - f105w - Outliers - RGB 3-color Images

```python
import numpy as np
import matplotlib.pyplot as plt
import sep
```

```python
from astropy.io import fits
from matplotlib import rcParams

%matplotlib inline

rcParams['figure.figsize'] = [10., 8.]
```

```python
#import fits data
hdu = fits.open('hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits')
hdu.info()

data_1 = hdu[0].data

hdu.close()
#print(type(data))
```

```python
# show the image
data_1 = data_1.byteswap().newbyteorder()
m, s = np.mean(data_1), np.std(data_1)

plt.imshow(data_1, interpolation='nearest', cmap='gray', vmin=m-s, vmax=m+s, origin='lower')
plt.colorbar();

plt.savefig('f105w ini.pdf',bbox_inches='tight',pad_inches=0,dpi=300,)
```

Background subtraction

```python
# measure a spatially varying background on the image
#data_1 = data_1.byteswap().newbyteorder()
bkg = sep.Background(data_1)
```

```python
# get a "global" mean and noise of the image background:
print(bkg.globalback)
print(bkg.globalrms)
```

```python
# evaluate background as 2-d array, same size as original image
bkg_image = bkg.back()
# bkg_image = np.array(bkg) # equivalent to above
```

```python
# show the background
plt.imshow(bkg_image, interpolation='nearest', cmap='gray', origin='lower')
plt.colorbar();
plt.savefig('f105w background.pdf',bbox_inches='tight',pad_inches=0,dpi=300,)
```

```python
# evaluate the background noise as 2-d array, same size as original image
bkg_rms = bkg.rms()
```

```python
# show the background noise
plt.imshow(bkg_rms, interpolation='nearest', cmap='gray', origin='lower')
plt.colorbar();
plt.savefig('f105w background noise.pdf',bbox_inches='tight',pad_inches=0,dpi=300,)
```

```python
# subtract the background
data_sub = data_1 - bkg
```

Object Detection

```python
#from matplotlib.patches import Ellipse

# plot background-subtracted image
fig, ax = plt.subplots()
m, s = np.mean(data_sub), np.std(data_sub)
im = ax.imshow(data_sub, interpolation='nearest', cmap='gray',
               vmin=m-s, vmax=m+s, origin='lower')

plt.savefig('f105w no ellipse plot.pdf',bbox_inches='tight',pad_inches=0,dpi=500,)
```

```python
# detect objects

#NOTE the 6.5 threshold
objects = sep.extract(data_sub, 6.5, err=bkg.globalrms)

# how many objects were detected
len(objects)
```

I spent some time playing with the threshold value, and comparing the extracted and marked image with the non extracted image. I did this after noticing "vignette noise" found on the edges of the image were confusing the sep.extraction function, marking high contrast grain as actual objects. My soloution was to change the threshold from 1.5 to 6.5, which removes the majority of the fake "noise objects" while preserving the majority of the real objects. While I realize by doing this, I am ignoring many of the fainter objects that sep.extract is no longer detecting, but on the flip size, the vast majority of objects being detected are "real objects" instead of the several thousand "noise objects" that were there before.

My final count of objects, using a threshold of 6.5 instead of the tutorial defualt of 1.5, is 2122.

The solution to removing all "noise objects" from the image edge while preserving the majority of real objects is to simply crop in a little, using area that is free from high contrast grain.

```python
from matplotlib.patches import Ellipse

# plot background-subtracted image
fig, ax = plt.subplots()
m, s = np.mean(data_sub), np.std(data_sub)
im = ax.imshow(data_sub, interpolation='nearest', cmap='gray',
               vmin=m-s, vmax=m+s, origin='lower')

# plot an ellipse for each object
for i in range(len(objects)):
    e = Ellipse(xy=(objects['x'][i], objects['y'][i]),
                width=8*objects['a'][i],
                height=8*objects['b'][i],
                angle=objects['theta'][i] * 180. / np.pi)
    e.set_facecolor('none')
    e.set_edgecolor('red')
    ax.add_artist(e)
plt.savefig('f105w final plot.pdf',bbox_inches='tight',pad_inches=0,dpi=500,)
```

^^^ NOTE: With exception to the visible surge on the bottom right edge, the vast majority of marked objects are "real objects", not "noise objects".

# Mean, Median, and Standard Deviation

```python
#checking mean median and STD of the flux data here.

flux_mean = np.mean(flux)
flux_median = np.median(flux)
flux_std = np.std(flux)

print(f'The flux mean is: {flux_mean}')
print(f'The flux median is: {flux_median}')
print(f'The flux standard deviation is: {flux_std}')
```

Presuming that these mean, median, and STD numbers are accurate, the median shows that the lower half of the objects are extremely low flux(brightness), no more than 0.2. In addition, the 2 outlers present in the distribution shift both the median and STD to the right in thier direction. Regarding the STD and outliers, in the case of the brightest object, the local Milky Way star, we can determine how many standard deviation away it is.

```python
#firstly, we determine the flux value of the excessively bright appearing star:
bright_star = np.amax(flux)
print(f'The flux value of the bright outlier star is {bright_star}')

#next, all we do is divide the flux value of the bright star by the value of one standard deviation:
bright_std_distance = bright_star / flux_std

print(f'The flux standard deviation is: {flux_std}')
print(f'The outlier bright star is "{bright_std_distance}-sigma"')
```

While the chaos of 43-sigma has the probability of some minuscule number not even worth calculating, in the context of where this data is coming from, it does make sense. The vast majority of objects were are detecting within the HUDF are extrememly distant and faint galaxies, possibly billions of light years away. Meanwhile, the hand full of outlier stars present in the HUDF are not far at all, no more than a few thousand light years perhaps. The way i see it, imagine trying to see a single burning candle several thousand miles away while someone is shining a flashlight directly at your eyes. Your probability of seeing that candle is roughly nonexistant, and likely two digit deviations away as well.

# HUDF 3 color image

```python
from astropy.io import fits
```

```python
#define file names:

fdata_f105w = "hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits"
fdata_f125w = "hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits"
fdata_f160w = "hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits"

#read in data
hdu_2 = fits.open(fdata_f105w)
hdu_3 = fits.open(fdata_f125w)
hdu_4 = fits.open(fdata_f160w)

#get the image data
data_2 = hdu_2[0].data
data_3 = hdu_3[0].data
data_4 = hdu_4[0].data

hdu_2.close()
hdu_4.close()
hdu_4.close()
```

```python
#f = plt.figure(figsize =(10,10))
#plt.imshow(data_2)
```

```python
def rescale_image(data):
    pdata_tmp = data.copy()
    m = np.nanmean(pdata_tmp)
    vplmin = m/2
    vpmin = np.log10(vplmin)
    vpmax = np.log10(m * 100.)
    pdata_tmp[pdata_tmp<vplmin] = vplmin
    pdata_tmp = np.log10(pdata_tmp)
    return pdata_tmp,vpmin,vpmax
```

```python
def fits_quicklook(data,fnx=10,fny=10):
    f = plt.figure(figsize=(fnx,fny))
    pdata_tmp,vpmin,vpmax = rescale_image(data)
    plt.imshow(pdata_tmp,vmin=vpmin,vmax=vpmax)
```

```python
fits_quicklook(data_2)
```

```python
#rescaled images
data_2_res, d2min, d2max = rescale_image(data_2)
data_3_res, d3min, d3max = rescale_image(data_3)
data_4_res, d4min, d4max = rescale_image(data_4)
```

```python
data_2_res[data_2_res<d2min] = d2min
data_2_res[data_2_res>d2max] = d2max

data_3_res[data_3_res<d3min] = d3min
data_3_res[data_3_res>d3max] = d3max

data_4_res[data_4_res<d4min] = d4min
data_4_res[data_4_res>d4max] = d4max
```

```python
rgb = np.zeros((data_2_res.shape[0], data_2_res.shape[1], 3))
rgb[:,:,0] = (data_2_res-d2min)/(d2max-d2min)
rgb[:,:,1] = (data_3_res-d3min)/(d3max-d3min)
rgb[:,:,2] = (data_4_res-d4min)/(d2max-d4min)
```

```python
plt.imshow(rgb)
plt.axis('off')
plt.savefig('HUDF RGB image.png',bbox_inches='tight',pad_inches=0,dpi=400,)
```
